```python
import os
print(os.getcwd())
import numpy as np
import scipy.special as sp
import seaborn as sns
import matplotlib
matplotlib.rcParams['pdf.fonttype'] = 42
matplotlib.rcParams['ps.fonttype'] = 42
import matplotlib.pyplot as plt
import scipy.io as sio
```

```python
pi = np.pi
def nmw212(t, n = 40, th = .1):
    b = (th * np.exp(-(th**2+t**2)/(4*t))*np.cosh(th/2))/np.sqrt(pi*t**3)
    c = ((1 + np.exp(-th))*sp.erfc((th-t)/np.sqrt(4*t))+(1+np.exp(th))*sp.erfc((th + t)/np.sqrt(4*t)))/2
    y = n*(1-c)**(n-1)*b
    return y

def nmw213(t, n = 40, th = .1):
    b = (th * np.exp(-(th**2+t**2)/(4*t))*np.cosh(th/2))/np.sqrt(pi*t**3)
    c = ((1 + np.exp(-th))*sp.erfc((th-t)/np.sqrt(4*t))+(1+np.exp(th))*sp.erfc((th + t)/np.sqrt(4*t)))/2
    y = n*np.exp((1-n)*c)*b
    return y

def nmw214(t, n=40, th = 0.1):
    b = (th * np.exp(-(th**2+t**2)/(4*t))*np.cosh(th/2))/np.sqrt(pi*t**3)
    y = b*n*np.exp(np.sqrt(4*t/pi)*(1-n)*np.exp(-th**2/(4*t))*(((2*th**2-4*t)/th**3)*np.cosh(th/2) - np.sinh(th/2)))
    return y

sns.set()
sns.set(font_scale=1.2)
sns.set_style('ticks')
red_colors = sns.color_palette("Reds_d",9)
blue_colors = sns.color_palette("Blues_d",9)
orange_colors = sns.color_palette("YlOrBr_d",9)
```

```python
# The First Decision Time Distribution for theta = 0.5, n = 40
homoN40 = sio.loadmat('z_0_1_Cooked_n40.mat')

time40 = np.linspace(0.0,0.002)
eq212n40 = [nmw212(t) for t in time40]
eq214n40 = [nmw214(t) for t in time40]
eq213n40 = [nmw213(t) for t in time40]

hist40 = homoN40['histTime']
sns.distplot(hist40, color = 'orange',label = 'Simulations')
plt.plot(time40,eq212n40, color='deepskyblue', lw=2, alpha=2, label = 'Eq 2.12') 
plt.plot(time40,eq213n40, color='darkcyan', lw=1.5, alpha=1, label = 'Eq 2.13') 
plt.plot(time40,eq214n40, color='rebeccapurple', lw=1.5, alpha=1,label = 'Eq 2.14') 

plt.legend(fontsize=12, frameon=False)
plt.xlabel('First Decision Time')
plt.title('n = 40')

sns.despine()
```
